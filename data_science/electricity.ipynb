```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from statsmodels.tsa.holtwinters import ExponentialSmoothing
```

```python
df = pd.read_csv('data/electricity.csv', index_col=0)
df.rename(columns={'X1': 'data'}, inplace=True)
df.head(5)
```

```python
plt.plot(df['data'])
plt.title('Time Series Data')
plt.xlabel('Time')
plt.ylabel('Value')
plt.grid(True)
plt.show()
```

```python
period = 7
model = ExponentialSmoothing(
    df['data'],
    trend='add',
    seasonal='add',
    seasonal_periods=period
)
fit = model.fit()

forecast = fit.forecast(period)

result = [str(int(round(x))) for x in forecast]
print(','.join(result))
```
