```python
import pandas as pd
```

```python
df = pd.read_csv('data/pulsar_stars.csv')
df.head()
```

```python
first_filter = (df.TG == 0) & (df.MIP >= 84.3515625) & (df.MIP <= 85.5234375)
second_filter = (df.TG == 1) & (df.MIP >= 47.4140625) & (df.MIP <= 52.9296875)

df_filtered = df[first_filter | second_filter]
```
