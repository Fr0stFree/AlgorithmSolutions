```python
import pandas as pd
import datetime as dt
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("./data/apartments.csv", delimiter=";", index_col="ID")
df.head(10)
```

```python
# remove 0 internet rows
df = df[df["INTERNET"] != 0]
```

```python
# add normalized columns
df["NORM_STOP_COUNT"] = (df["STOP_COUNT"].max() - df["STOP_COUNT"]) / (df["STOP_COUNT"].max() - df["STOP_COUNT"].min())
df["NORM_COST"] = (df["COST"].max() - df["COST"]) / (df["COST"].max() - df["COST"].min())
df["NORM_DISTANCE"] = (df["DISTANCE"].max() - df["DISTANCE"]) / (df["DISTANCE"].max() - df["DISTANCE"].min())
df["NORM_FITNESS"] = (df["FITNESS"].max() - df["FITNESS"]) / (df["FITNESS"].max() - df["FITNESS"].min())
df["NORM_INTERNET"] = (df["INTERNET"] - df["INTERNET"].min()) / (df["INTERNET"].max() - df["INTERNET"].min())
df.head(5)
```

```python
df["SCORE"] = (
    df["NORM_STOP_COUNT"] +
    df["NORM_COST"] +
    df["NORM_DISTANCE"] +
    df["NORM_FITNESS"] * 0.5 +
    df["NORM_INTERNET"] * -0.2
)
df = df.sort_values(by="SCORE", ascending=False)
print(df[['DISTANCE', 'STOP_COUNT', 'COST', 'FITNESS', 'INTERNET', 'SCORE']].head())
```
